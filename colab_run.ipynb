### Step 3: Set up Ngrok

```python
from pyngrok import ngrok

# Replace 'YOUR_NGROK_AUTHTOKEN' with your actual token
NGROK_AUTH_TOKEN = "YOUR_NGROK_AUTHTOKEN"
ngrok.set_auth_token(NGROK_AUTH_TOKEN)
```

### Step 4: Run the Application
This will start the Flask server and provide a public URL where you can access your app.

```python
import os
from pyngrok import ngrok

# Open a HTTP tunnel on default port 5000
public_url = ngrok.connect(5000).public_url
print(f" * Flask App is live at: {public_url}")

# Run the app
!python app.py
```
